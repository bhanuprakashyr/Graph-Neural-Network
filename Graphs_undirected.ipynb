# Learn Graph using Networkx  
NetworkX is a powerful Python library for creating, manipulating, and analyzing complex networks and graphs. It is widely used for modeling real-world systems such as social networks, biological pathways, transportation systems, and more, enabling tasks like pathfinding, centrality analysis, and visualization.

### Graph Types
- **Undirected Graphs (nx.Graph)**: Edges are bidirectional, representing symmetric relationships.
- **Directed Graphs (nx.DiGraph)**: Edges have direction, indicating one-way relationships.
- **Weighted Graphs**: Edges can have weights (e.g., distances or costs), applicable to both undirected and directed graphs.

## Undirected Graphs (nx.Graph)
Undirected graphs treat edges as bidirectional.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

### Addition Methods
- `add_node(node, **attr)`: Adds a single node with optional attributes.
- `add_nodes_from(nodes, **attr)`: Adds multiple nodes from an iterable, with optional attributes.
- `add_edge(u, v, **attr)`: Adds an undirected edge between nodes u and v, with optional attributes (e.g., weight).
- `add_edges_from(edges, **attr)`: Adds multiple undirected edges from an iterable of (u, v) pairs, with optional attributes.
- `add_weighted_edges_from(edges)`: Adds multiple weighted edges from an iterable of (u, v, weight) tuples.

```python
# Create a graph representing major cities in Karnataka, India
G = nx.Graph()

# Add a node with an attribute
G.add_node('Mysore')
G.nodes['Mysore']['population'] = '1M'

# Add cities as nodes with attributes in a single go
cities_with_attrs = [
    ('Bangalore', {'population': '10M'}),
    ('Mysore', {'population': '1M'}),
    ('Mangalore', {'population': '0.5M'}),
    ('Hubli', {'population': '1.2M'}),
    ('Belgaum', {'population': '0.6M'}),
    ('Gulbarga', {'population': '0.5M'}),
    ('Davangere', {'population': '0.4M'})
]
G.add_nodes_from(cities_with_attrs)

# Add a single edge with weight
G.add_edge('Davangere', 'Bangalore', weight=250)

# Add edges with weights in a single go
weighted_edges = [
    ('Bangalore', 'Mysore', 150),
    ('Bangalore', 'Hubli', 400),
    ('Mysore', 'Mangalore', 250),
    ('Hubli', 'Belgaum', 100),
    ('Belgaum', 'Gulbarga', 200),
    ('Gulbarga', 'Davangere', 180),
    ('Davangere', 'Bangalore', 250)
]
G.add_weighted_edges_from(weighted_edges)

# Print nodes and edges
print("Cities (Nodes):", list(G.nodes()))
print("Road Connections (Edges):", list(G.edges()))

# Simple visualization
nx.draw(G, with_labels=True)
plt.title("Major Cities in Karnataka Graph - simple visualization")
plt.show()

# Visualize the graph
pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(10,8))

# Draw nodes and edges
nx.draw(G, pos, with_labels=False, node_color='lightgreen', node_size=2000, edge_color='gray')

# Create labels for nodes including attributes
node_labels = {node: f"{node}\n({G.nodes[node]['population']})" for node in G.nodes()}
nx.draw_networkx_labels(G, pos, labels=node_labels, font_size=9)

# Display edge weights
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels, font_size=8)

plt.title("Major Cities in Karnataka - Road Network with Populations")
plt.show()
```

### Modification Methods
- `subgraph(nodes)`: Returns a subgraph induced by the given nodes.

```python
subG = G.subgraph(['Bangalore', 'Gulbarga','Davangere'])
print(subG.edges())
print(subG.nodes(data=True))
nx.draw(subG, with_labels=True)
plt.show()

# We can see the attributes are retained in the subgraph
```

### Deletion Methods
- `remove_node(node)`: Removes a node and all its incident edges.
- `remove_nodes_from(nodes)`: Removes multiple nodes and their incident edges.
- `remove_edge(u, v)`: Removes the undirected edge between u and v.
- `remove_edges_from(edges)`: Removes multiple edges from an iterable of (u, v) pairs.

```python
# Remove single node
G.remove_node('Mangalore')
nx.draw(G, with_labels=True)
plt.show()
```

```python
# Remove Multiple Nodes
G.remove_nodes_from(['Belgaum','Hubli'])
nx.draw(G, with_labels=True)
plt.show()
```

```python
# Remove single Edge
G.remove_edge('Davangere','Bangalore')
nx.draw(G, with_labels=True)
plt.show()
```

```python
# Remove Multiple Edge
G.remove_edges_from([('Davangere','Bangalore'),('Gulbarga','Davangere')])
nx.draw(G, with_labels=True)
plt.show()
```

### Modification Methods
- `clear()`: Removes all nodes and edges.
- `copy()`: Returns a shallow copy of the graph.
- `update(other)`: Updates the graph with nodes/edges from another graph.

```python
G.clear()
nx.draw(G, with_labels=True)
plt.show()
```

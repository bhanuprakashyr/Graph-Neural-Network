## Directed Graphs (nx.DiGraph)
Directed graphs have edges with direction (from u to v).

### Addition Methods
- `add_node(node, **attr)`: Same as undirected.
- `add_nodes_from(nodes, **attr)`: Same as undirected.
- `add_edge(u, v, **attr)`: Adds a directed edge from u to v, with optional attributes.
- `add_edges_from(edges, **attr)`: Adds multiple directed edges from an iterable of (u, v) pairs.
- `add_weighted_edges_from(edges)`: Same as undirected, but edges are directed.

### Deletion Methods
- `remove_node(node)`: Same as undirected, removes node and all in/out edges.
- `remove_nodes_from(nodes)`: Same as undirected.
- `remove_edge(u, v)`: Removes the directed edge from u to v.
- `remove_edges_from(edges)`: Removes multiple directed edges.

### Modification Methods
- `clear()`: Same as undirected.
- `subgraph(nodes)`: Same as undirected.
- `copy()`: Same as undirected.
- `update(other)`: Same as undirected.
- `reverse()`: Returns a new graph with all edges reversed.
